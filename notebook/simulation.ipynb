This notebook is about using EM algorithm on two different simulated data.

```python
# env: CASIA2024-final
```

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
from functions.generate_simulated_data import Generate_Simulated_Data
from functions.EM_poisson_nmf import EM_poisson_nmf
from functions.match_EM_real import match_EM_real
```

```python
n = 100
m = 400
K = 6
max_iter = 1000
random_seed = 0
```

Simulation 1

```python
# generate simulated data
data1 = Generate_Simulated_Data()
X1, L1_real, F1_real = data1.generate_data(n, m, K, random_seed, choice=1)
```

```python
# fit the model
model1 = EM_poisson_nmf()
L1_em, F1_em, loglikelihood_list1 = model1.fit(X1, K, max_iter, random_seed)
```

```python
plt.figure()
plt.plot(loglikelihood_list1[-1] - loglikelihood_list1)
plt.xlabel('iteration')
plt.ylabel('distance from the best loglikelihood')
# plt.title('loglikelihood vs iteration')
plt.show()
```

```python
# Find the match between the EM labels and the real labels
match_em_real_list1 = match_EM_real(L1_em, L1_real)

legend1 = []
for k in range(K):
    legend1 += [str(k+1)]

# plot
plt.figure()
plt.plot(L1_real, L1_em[:, match_em_real_list1], 'o')
plt.xlabel('L_real')
plt.ylabel('L_em')
# plt.title('L_real vs L_em')
plt.legend(legend1)
plt.show()
```
